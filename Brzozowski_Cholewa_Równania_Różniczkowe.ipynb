## 1.1 Model Kermacka i McKendicka (SIR) 
Jest to jeden z podstawowych wzorców biologii matematycznej, na nim oparliśmy dalsze rozważania, dostosowując go odpowiednio do danej choroby. 
Model prezentuje się następująco: 
$$
\begin{cases}
S'(t)=-\frac{\alpha}{N}S(t)I(t),\\
I'(t)=\frac{\alpha}{N} S(t)I(t) -\beta I(t),\\
R'(t)=\beta I(t),
\end{cases}
$$
gdzie $S$ (ang. susceptible) to zbiorowisko zdrowych i podatnych na chorobę jednostek, $I$ (ang. infected) to zbiorowisko zainfekowanych jednostek,
$R$ (ang. resistant/remove) to zbiorowsiko odpornych lub usuniętych jednostek, a współczyniki $\alpha$ i $\beta$ to odpowiednio "zakażenia" i "ozdroweinia" (przejścia z grupy $I$ do $R$). Natomiast $N$ to stała liczba ludności.

W tym wzorcu musimy przyjąć następujące założenia:
- jednostka zdrowa, która się zarazi, natychmiast przenosi się do grupy $I$ z możliwością zarażenia innych zdrowych osobników,
- jednostka zarażona przenosi się do grupy $R$ w wyniku śmierci lub ozdrowienia (nabywa trwała odporność),
- jednostka zarażona, zakaża innych średnio w czasie $\frac{\alpha}{N} S$,
- choroba jest przenoszona przez bezpośredni kontakt oraz pomijamy wszystkie czynniki nie związane z epidemią, przyjmując, że liczba ludności pozostaje cały czas taka sama $N = S + I + R$.

## 1.2 Analiza Modelu SIR
Dokonując analizy modelów epidemiologicznych, głównie chcemy wiedzieć w jakich okolicznościach może i czy w ogóle dojdzie do wybuchu epidemii i jaki będzię jej przebieg. Do takiej analizy możemy uprościć nasz układ równań, pomijając w nim równanie grupy $R$, ponieważ nie występuję w  żadnym z pozostałych równań, z wyjątkiem rówanania opisującego zachowanie tej grupy, ale nadal możemy badać ów zbiór, analizować korzystając z założenia $N=S+I+R$. 

Zatem analizować będziemy dwa pierwsze równania:
$$
\begin{cases}
S'(t)=-\gamma S(t)I(t),\\
I'(t)=\gamma S(t)I(t) -\beta I(t).\\
\end{cases}
$$
Dla łatwiejszej analizy wprowadziliśmy $\gamma=\frac{\alpha}{N}$.

Na dobry początek zajmijmy się pierwszym równaniem. Zauważyć możemy w łatwy sposób, że $\forall t$ pochodna $S'(t) \leq 0$. Z czego możemy wnioskować, że funkcja $S(t)$ jest nierosnąca. Zastanówmy się zatem nad monotonicznością funkcji $I(t)$, zbadajmy jej pochodną.
Zapiszmy drugie z równań w poniższy sposób:
$$
\begin{split}
I'(t) & = \gamma S(t)I(t) -\beta I(t),\\
      & = (\gamma S(t) - \beta)I(t).
\end{split}
$$
Z założeń modelu, wiemy, że funkcje opisujące każdą z grup są nieujemne. Wtedy przy analizie znaku pochodnej $I'(t)$ zwrócimy uwagę na część: $\gamma S(t) - \beta$.

Zatem mamy dwie możliwości:
- a) $\gamma S(t) - \beta > 0 \iff S(t) > \frac{\beta}{\gamma}  \implies I'(t) > 0$,
- b) $\gamma S(t) - \beta \leq 0 \iff S(t) \leq \frac{\beta}{\gamma}  \implies I'(t) \leq 0 $.

Co interpretujemy, że od pewnego $t$ wartości funkcj $S(t) > \frac{\beta}{\gamma}$, więc mamy do czynienia ze wzrostem liczby osób zakażonych (tak jak wyżej  pokazaliśmy $I'(t) >0$), zatem widoczny będzię wybuch epidemii. Natomiat dla pewnych wartości funkcji mniejszych od tego pewnego $t$ (tak jak wyżej pokazaliśmy $I'(t) \leq 0$) , mamy zatem do czynienia ze spadkiem liczby osób podatnych, co powoduje zagaszenie się epiedemii, więc nie nastąpi jej wybuch. Liczbę $\frac{\beta}{\gamma}$ możemy interpretować jako wskaźnik progu, kiedy wybucha epidemia. (W literaturze związanej z tym tematem, wskaźnik ten określają jako minimalną liczbę osobników podatnych, niezbędnych do rozwoju epidemii). Co więcej, możemy założyć, że wcześniej wspominana pewna chwila $t$ następuję w momencie, gdy $t=0$, wtedy możemy zastąpić funkjcę $S$ jej założeniem początkowym $S(0)=S_0$.

Mamy więc: 
- c) $\gamma S_{0} - \beta > 0 \iff S_{0} > \frac{\beta}{\gamma}  \implies$ możliwość a) $\forall t \geq 0 $ ,
- d) $\gamma S_{0} - \beta \leq 0 \iff S_{0} \leq \frac{\beta}{\gamma}  \implies$ możliwość b) $\forall t \geq 0 $ .


Podsumowując twierdzimy, że rozwój i wybuch epidemii zależy od warunku początkowego $S(0)=S_{0}$ i jego wartości przyrównanej do stałego ułamka $\frac{\beta}{\gamma}$.

W paragrafie 1.3 postaramy się udowodnić prawdziwość naszego wniosku, rozwiącując model numerycznie. 

## 1.3 Rozwiązanie Numeryczne Modelu SIR
Zastanówmy się teraz nad rozwiązaniem numerycznym naszego układu równań, jaki prezentuje model SIR. Przypomnijmy zagadnienia Cauchy'ego $S(0)=S_{0}$, $I(0)=I_{0}$ i $R(0)=R_{0}$. Do znalezienia rozwiązania posłużymy się metodą Eulera. Aby ułatwić sobie sprawę zamieńmy nasz układ z punktu 1.1 na układ :
$$
\begin{cases}
S'(t)=-\gamma S(t)I(t),\\
I'(t)=\gamma S(t)I(t) -\beta I(t),\\
R'(t)=\beta I(t).
\end{cases}
$$
Zauważmy, że stała $\frac{\alpha}{N} = \gamma$ (poza tą jedną zmianną, układy nie różnią się niczym).

W chwili $t=0$ nasz układ równań wygląda następująco:
$$
\begin{cases}
S'(0)=-\gamma S(0)I(0) = -\gamma S_{0}I_{0} ,\\
I'(0)=\gamma S(0)I(0) -\beta I(0) = \gamma S_{0}I_{0} -\beta I_{0}  ,\\
R'(0)=\beta I(0)=\beta I_{0}.
\end{cases}
$$
Natomiast, gdy $t=1$, krok w przód mamy:
$$
\begin{cases}
S(1)\approx S(0) + S'(0)= S_{0} -\gamma S_{0}I_{0} ,\\
I(1)\approx I(0) + I'(0) = I_{0} + \gamma S_{0}I_{0} -\beta I_{0}  ,\\
R(1) \approx R(0) + R'(0) = R_{0} + \beta I_{0}.
\end{cases}
$$
Następne kroki w przód wynikają z poniższego schematu: 
- Do znalezienia rozwiązania dla chwili $t=2$ skorzystamy z wyznaczonych już wartości $S(1), I(1), R(1)$ i również z wartości $S'(1), I'(1), R'(1)$, które pozyskamy stosując działania takie same jak przy pozyskaniu $S'(0), I'(0), R'(0)$.
- Stąd mamy już $S(2), I(2), R(2)$
- Cały schemat powtarzamy.

Wniosek, prowadzi nas do ogólniejszego opisu kroku w przód w metodzie Eulera:
$$
\begin{cases}
S(i+1)\approx S(i) + S'(i)= S_{i} -\gamma S_{i}I_{i} ,\\
I(i+1)\approx I(i) + I'(i) = I_{i} + \gamma S_{i}I_{i} -\beta I_{i}  ,\\
R(i+1)\approx  R(i) + R'(i) = R_{i} - \beta I_{i}.
\end{cases}
$$

Za pomocą powyższego układu, zdecydowaliśmy się przedstawić wyniki tej metody w arkuszu Excel i za pomocą programu napisanego w języku Python.  

Zacznijmy od rozwiązanaia problemu w Excelu (załączony Symulacja(SIR).xlsx) ,  główną zaletą przeprowadzenia rozwiązania w tym programie jest to, że możemy w szybki sposób otrzymać symulacje, na długi okres czasu. Co pozwala na analzię zachowań funkcji na różnych okresach czasowych.


Dla lepszego efektu, stworzyliśmy również symulacje w Pythonie:

```python
import pylab
import numpy as np

def SIR(S0,I0,R0,gamma,beta):
    S=[]
    I=[]
    R=[]
    T=[]
    t = 0  
    S.append(S0)
    I.append(I0)
    R.append(R0)
    while t <= 100:
        S.append(S[t] - gamma*S[t]*I[t])
        I.append(I[t] + gamma*S[t]*I[t] - beta*I[t])
        R.append(R[t]+beta*I[t])
        t=t+1
    S.pop(0)
    I.pop(0)
    R.pop(0)
    i = np.arange(0,101)

    pylab.plot(i,S,label='S')
    pylab.plot(i,I,label='I')
    pylab.plot(i,R,label='R')
    pylab.title('SIR')
    pylab.xlabel('Czas')
    pylab.ylabel('Populacja')
    pylab.legend()
    pylab.show()
    print('S(0) wynosi: ', S0, ' a Beta/Gamma wynosi:', beta/gamma)

print(SIR(9, 1, 0, 0.015, 0.05))
print(SIR(9, 1, 0, 0.005, 0.05))
```

Dzięki symulacjom i rozwiązaniu numerycznemu układu, potwierdziliśmy prawdziwość wniosków z paragrafu 1.1.

# 1.4 Model SEIR
W tym paragrafie chcemy pokazać, że możemy edytować modele. Jak wpomnieliśmy wcześniej, rozwój chorób wprowadził zapotrzebowanie na edycje istniejących już modeli. Tak samo podstawowy model SIR można dostosować, do chorób nie spełniających wcześniej wspomnianych warunków. Przykładowo, możemy wprowadzić nowe współczynniki i kolejną grupę ludzi $E$ (ang. exposed) to zbiorowisko zarażonych, w których choroba inkubuje. Wtedy taki model wyraża się jako:
$$
\begin{cases}
S'(t)=-\frac{\alpha}{N}S(t)I(t),\\
E'(t)= \frac{\alpha}{N}S(t)I(t)-\sigma E(t),\\
I'(t)=\sigma E(t) -\gamma I(t),\\
R'(t)=\gamma I(t),
\end{cases}
$$
gdzie współczynniki $\alpha$, $\sigma$ i $\gamma$ to odpowiednio "zarażenia" (tak jak w poprzednim modelu), "wylęgania" choroby i "ozdroweinia" (przejścia z grupy $I$ na $R$), a reszta funkcji i stała $N$ oznacza to samo co w modelu SIR.

Najlepszym sposobem na przedstawienie tego modelu, będzie jego symulacja (rozwiązanie numeryczne metodą Eulera). Pisząc program w języku python możemy sprawdzić, jak w zależności od czasu zachowują się grupy w Modelu SEIR. Poniżej przeprowadziliśmy badanie na takiej samej grupie ludzi i o takich samych współczynnikach jak w paragrafie 1.3, jedynie zmienialiśmy stałą odpowiadającą za wylęganie się choroby.

```python
import random
import pylab
import numpy as np
def SEIR(S0,E0,I0,R0,alfa,sigma,gamma):
    N=S0+E0+I0+R0
    S=[]
    E=[]
    I=[]
    R=[]
    t = 0  
    S.append(S0)
    E.append(E0)
    I.append(I0)
    R.append(R0)
    while t <= 100:
        S.append(S[t] - (alfa*S[t]*I[t])/N)
        E.append(E[t]+(alfa*I[t]*S[t])/N - sigma*E[t])
        I.append(I[t] + sigma*E[t] - gamma*I[t])
        R.append(R[t]+gamma*I[t])
        t=t+1
    S.pop(0)
    E.pop(0)
    I.pop(0)
    R.pop(0)
    i = np.arange(0,101)

    pylab.plot(i,S,label=('S'))
    pylab.plot(i,E,label=('E'))
    pylab.plot(i,I,label=('I'))
    pylab.plot(i,R,label=('R'))
    pylab.title('SEIR')
    pylab.xlabel('Czas')
    pylab.ylabel('Populacja')
    pylab.legend()
    pylab.show()
    print('Wspolczynik inkubacji wynosi: ', sigma)

print(SEIR(9, 1, 0, 0,  1.5, 0.19, 0.05))
print(SEIR(9, 1, 0, 0, 1.5, 0.045, 0.05))
print(SEIR(9, 1, 0, 0, 1.5, 0.5, 0.05))
print(SEIR(9, 1, 0, 0, 0.05,0.19, 0.05))
```

Widzimy, że w tym modelu, ważny jest również współczynnik inkubacji, ponieważ wpływa na moment i intensywność wybuchu epidemii. Celem tego podpunktu, było pokazanie, że możemy edytować podstawowy model Kermacka i McKendricka. Zgodnie z założeniem pracy przejdźmy do modelowania chorób wenerycznych.

# 2.1 Specyfika Chorób Wenerycznych
Poprzedni paragraf pokazał nam, że możemy edytować istniejące już modele, tak aby dopasować je do różnych chorób. Przypadłości przenoszone drogą płciową są jednak dość specyficznymi dolegliwościami, od zwykłych chorób odróznia je między innymi to, że:
- liczba całej grupy jest stworzona tylko z osób aktywnych seksualnie,  
- osoby po przejściu z grupy $I$ do $R$, nie nabywają odporności,
- często chore jednostki, aż do zaawonsowanego stadium choroby nie wykazują żadnych objawów.

Musimy mieć więc na względzie, powyżej wymienione różnice dostosowując model Kermacka i McKendricka do danej choroby.

# 2.2 Podstawowy Model Chorób Wenerycznych
Podporządkujmy teraz shcemat SIR do dowolnej choroby wenerycznej. Dokładamy jednak dodatkowe, założenia, czyli:
- cała grupa składa się z dwóch niezmiennych liczbowo podgrup podzielnych na płeć: Kobiet i Mężczyzn,
- rozpatrujemy kontakty heteroseksualne, zakażenie może nastapić tylko podczas kontaktu jednostek z dwóch różnych grup.

Stworzyliśmy nasz model opierając go na rodzaju budowy schematów SIR, jak wcześniej pokazaliśmy, jesteśmy w stanie edytować wzory. Jak podkreśliliśmy w części 2.1, osoby nie nabywają odporności, co oznacza, że nadal są podatne na zakażenie, możemy zatem uprościć wzorzec i pominąć w nim grupę $R$. Wtedy zastosowalibyśmy model SI, ale typowe dla niego jest to, że nie ma możliwości wyzdrowienia, a w naszym możliwe jest pokonanie choroby. Zatem skorzystaliśmy z naturalnego rozszerzenia wzorca SI, a dokładniej stworzyliśmy wzorzec SIS. Będzię on działał tak jak na poniższym schemacie: 

![SCHEMAT](schemat.png)

Nasz model prezentuje się następująco:
$$
\begin{cases}
S'_{K}(t)= -\frac{\alpha_{K}}{N_{K}}S_{K}(t)I_{M}(t) + \beta_{K} I_{K}(t),\\
I'_{K}(t)=\frac{\alpha_{K}}{N_{K}} S_{K}(t)I_{M}(t) -\beta_{K} I_{K}(t),\\
S'_{M}(t)= -\frac{\alpha_{M}}{N_{M}}S_{M}(t)I_{K}(t) + \beta_{M} I_{M}(t),\\
I'_{M}(t)=\frac{\alpha_{M}}{N_{M}} S_{M}(t)I_{K}(t) -\beta_{M} I_{M}(t),\\
\end{cases}
$$
gdzie dolny indeks oznacza: $K$ Kobiet, $M$ Mężczyzn. Wtedy, wspólczynniki $\alpha_{K}$ i $\alpha_{M}$ oznaczają "zakażenie" dla odpowiedniej grupy, a współczyniki $\beta_{K}$ i $\beta_{M}$ "ozdrowienie".


Do łatwiejszej analizy modelu podstawmy $\gamma_{M}=\frac{\alpha_{M}}{N_{M}}$ i $\gamma_{K}=\frac{\alpha_{M}}{N_{M}}$.
Wtedy model prezentuje się następująco:
$$
\begin{cases}
S'_{K}(t)= -\gamma_{K}S_{K}(t)I_{M}(t) + \beta_{K} I_{K}(t),\\
I'_{K}(t)=\gamma_{K} S_{K}(t)I_{M}(t) -\beta_{K} I_{K}(t),\\
S'_{M}(t)= - \gamma_{M} S_{M}(t)I_{K}(t) + \beta_{M} I_{M}(t),\\
I'_{M}(t)=\gamma_{M} S_{M}(t)I_{K}(t) -\beta_{M} I_{M}(t),\\
\end{cases}
$$

```python
import pylab
import numpy as np

def SIRcw(S0M,S0K,I0M,I0K,gammaM,gammaK,betaM,betaK):
    SM=[]
    IM=[]
    SK=[]
    IK=[]
    t = 0  
    SM.append(S0M)
    SK.append(S0K)
    IM.append(I0M)
    IK.append(I0K)
    while t <= 100:
        SM.append(SM[t] - gammaM*SM[t]*IK[t]+betaM*IM[t])
        SK.append(SK[t] - gammaK*SK[t]*IM[t]+betaK*IK[t])
        IM.append(IM[t] + gammaM*SM[t]*IK[t] - betaM*IM[t])
        IK.append(IK[t] + gammaK*SK[t]*IM[t] - betaK*IK[t])
        t=t+1
    SM.pop(0)
    IM.pop(0)
    SK.pop(0)
    IK.pop(0)
    i = np.arange(0,101)

    pylab.plot(i,SM,label='SM')
    pylab.plot(i,IM,label='IM')
    pylab.plot(i,SK,label='SK')
    pylab.plot(i,IK,label='IK')
    pylab.title('Model Choroby Wenerycznej')
    pylab.xlabel('Czas')
    pylab.ylabel('Populacja')
    pylab.legend()
    pylab.show()
print(SIRcw(9,9,1,1,0.05,0.5, 1.25,0.05))
print(SIRcw(9,9,1,1,0.045,0.035, 0.51, 0.01))
print(SIRcw(9,9,1,1,0.035,0.06,1.25,0.05))
print(SIRcw(9,9,1,1,0.035,0.3,0.25,0.05))
print(SIRcw(9,9,1,1,0.05,0.5, 0.8,0.05))
print(SIRcw(9,9,1,1,0.098,0.115,0.25,0.5))####
print(SIRcw(9,9,1,1,0.098,0.115, 0.9,0.5))
print(SIRcw(999,999,1,1,0.00098,0.00115,0.5,0.5))
```

Jak widzimy otrzymujemy różnego typu wykresy, wzięliśmy dowolną chorobę weneryczną, zatem ciężko odpowiedzieć nam czego naprawdę oczekujemy od symulacji. Jednak, warto zauważyc, że tak jak założyliśmy, dwie grupy oddziałują na siebie symetrycznie, więc model jak najbardziej spełnia nasze oczekiwania.

## 2.3 Model rozwoju HIV/AIDS

Zapiszmy teraz układ równań prezentujący zmiany liczebności poszczególnych grup na przestrzeni czasu:
$$
\begin{cases}
X'(t)= B - \mu X(t) - c \alpha X(t),\\
Y'(t)=c \alpha X(t) - (v + \mu )Y(t),\\
A'(t)=pvY(t) - (d + \mu )A(t),\\
\alpha = \beta \frac{Y(t)}{X(t)},\\
\end{cases}
$$
gdzie $Y'(t)$ możemy zapisać w alternatywnej wersji:
$Y'(t) \approx (\beta c-v-\mu)Y(t) \approx v(\frac{\beta c}{v}-1)Y(t)$, gdyż $v \gg \mu$$.


## 2.4 Analiza modelu HIV/AIDS 
Wprowadźmy założenia początkowe $X(0)=X_{0}$, $Y(0)=Y_{0}$ i $A(0)=A_{0}=0$

Zacznijmy od analizy wzoru: 
$$
Y'(t) \approx v(\frac{\beta c}{v}-1)Y(t).
$$
Badając znak pochodnej, możemy dojść do wniosku, że epidemia nastąpi w momencie, gdy $\frac{\beta c}{v} -1 >0 \implies \frac{\beta c}{v} > 1$

Następnie rozwiążemy równanie na liczbę zarażonych w zależności od czasu:
$$
Y'(t) = v(\frac{\beta c}{v} -1)Y(t) ,\\
Y'(t) - v(\frac{\beta c}{v} -1)Y(t) = 0.
$$

Obliczamy współczynnik całkujący:
$$
\eta = e^{-\int v(\frac{\beta c}{v} -1)dt} = e^{-v(\frac{\beta c}{v} -1)t}.
$$

Mnożymy obustronnie i otrzymujemy:
$$
Y'(t) e^{-v(\frac{\beta c}{v} -1)t} - [-v(\frac{\beta c}{v} -1)e^{-v(\frac{\ betac}{v} -1)t}] = 0,\\
(Y(t) e^{-v(\frac{\beta c}{v} -1)t})' = 0.
$$

Następnie, całukując obustronie: 
$$
Y(t) e^{-v(\frac{\beta c}{v} -1)t} - Y(0) e^{-v(\frac{βc}{v} -1) 0}= 0, \\
Y(t)e^{-v(\frac{\beta c}{v} -1)t} = Y_{0}, \\
Y(t) = Y_{0} e^{v(\frac{\beta c}{v} -1)t}.
$$

Znając $Y(t)$ możemy podstawić uzyskany wynik do równania opisującego rozwój liczebności pacjentów chorych na HIV:
$$
A'(t)=pvY(t) - (d+mu)A(t),\\
A'(t)=pvY_{0} e^{v(\frac{\beta c}{v} -1)t} - (d+ \mu )A,\\
A'(t) = (d+\mu)A(t) = pvY_{0} e^{v(\frac{\beta c}{v} -1)t}.
$$

Obliczamy współczynnik całkujący:
$$
\eta = e^{\int(d+\mu)dt} = e^{(d+\mu)t}.
$$

Mnożymy obustronie:
$$
A'(t) e^{(d+\mu )t}+(d+\mu)A(t)e^{(d+\mu)t} = pvY_{0}e^{(d+\mu)t}e^{v(\frac{\beta c}{v} -1)t},\\
(A(t) e^{(d+\mu)t})' = pvY_{0} e^{(d+\mu )t}e^{v(\frac{\beta c}{v} -1)t}.\\
$$

Całkujemy: 
$$
A(t) e^{(d+ \mu)t} +  A_{0} = \int pv Y_{0}e^{(d+\mu)t} + v(\frac{\beta c}{v}-1)dt,
$$
stosując podstawienie:
$$
x = (d+\mu)t + v(\frac{\beta c}{v} -1)t,\\
dt = \frac{dx}{(d+\mu )+ v(\frac{\beta c}{v} -1)},
$$
mamy:
$$
A(t)e^{(d+ \mu )t}+ A_{0} = pvY_{0}\int{\frac{e^{x}dx}{(d+\mu)+ v(\frac{\beta c}{v} -1)}} = \frac{pvY_{0} e^{(d+\mu )t + v(\frac{\beta c}{v} -1)t}-1}{(d+\mu)+ v(\frac{\beta c}{v} -1)}.\\
$$

Otrzymujemy zatem:
$$
A(t) = -\frac{A_{0}}{e^{(d+\mu)t}} + \frac{pvY_{0}e^{(d+\mu)t + v(\frac{\beta c}{v} -1)t}-e^{-(d+\mu)t}}{(d+\mu)+ v(\frac{\beta c}{v} -1)}.
$$

Ostatecznie odwołując się do początkowych założeń $A(0)=A_{0}=0$ dostajemy:

$$
A(t) =\frac{pvY_{0} e^{(d+\mu)t + v(\frac{\beta c}{v} -1)t}-e^{-(d+ \mu)t}}{(d+\mu)+ v(\frac{\beta c}{v} -1)} .
$$

Rozwiążmy jeszcze pierwsze równanie:
$$
X'(t)= B - \mu X(t) -c\alpha X(t),\\
X'(t) + (\mu +c\alpha)X(t)=B.
$$

Wyliczamy czynnik całkujący:
$$
\eta = e^{\int(\mu + ca)dt} = e^{(\mu + ca)t}.
$$

Mnożymy i całkujemy, otrzymujemy wtedy:
$$
X(t)= \frac{Bt+ C}{e^{(\mu + ac)t}}.
$$

Korzystając z założeń początkowych ostatecznie:
$$
X(t)= \frac{Bt+ X_{0}}{e^{(\mu + ac)t}}.
$$

Podsumowując, rozwiązanie analityczne naszego modelu wygląda następująco:
$$
\begin{cases}
X(t)=\frac{Bt+ X_{0}}{e^{(\mu + ac)t}},\\
Y(t)=Y_{0} e^{v(\frac{βc}{v} -1)t},\\
A(t)=\frac{pvY_{0} e^{(d+\mu)t + v(\frac{\beta c}{v} -1)t}-e^{-(d+ \mu)t}}{(d+\mu)+ v(\frac{\beta c}{v} -1)},\\
\alpha = \beta \frac{Y(t)}{X(t)}.
\end{cases}
$$

## 2.5 Rozwiązanie Numeryczne modelu HIV/AIDS
Użyjemy wcześniej już stosowanej przez nas metody Eulera. 


```python
import random
import pylab
import numpy as np
from math import exp 
def HIV(X0,Y0,A0,B,mi,beta,c,d,p,v):
   
    X=[]
    Y=[]
    A=[]
    t = 0
    N=X0+Y0+A0
    alfa=((beta*Y0)/X0)
    X.append(X0)
    Y.append(Y0)
    A.append(A0)

    while t <= 100:
        
        X.append(X[t] + B -(mi+c*alfa)*X[t])#(B*t + X0)/(exp((mi+alfa*c)*t)))  
        Y.append(Y[t] + c*(alfa)*X[t] - (v+mi)*Y[t])   #) Y0*exp((v*t*((beta*c)/v -1))))
        A.append(A[t] + p*(v+mi)*Y[t] - (d+ mi)*A[t])# (p*v*Y0*exp((d+mi)*t+v*t*((beta*c)/v -1))-exp(-t*(d+mi)))/((d+mi)+v*((beta*c)/v-1))) 
        t=t+1
    X.pop(0)
    Y.pop(0)
    A.pop(0)
    i = np.arange(0,101)

    
    pylab.plot(i,X,label=('X'))
    pylab.plot(i,Y,label=('Y'))
    pylab.plot(i,A,label=('A'))
    #pylab.plot(i,alfa)
    pylab.title('Symulacja Choroby HIV/AIDS')
    pylab.xlabel('Czas')
    pylab.ylabel('Populacja')
    pylab.legend()
    pylab.show()
    print('Beta*c/v:', beta*c/v, 'współczynnik konwersji przejścia z grupy Y do A:', v )
    
```

Rozważmy kilka sytuacji rozwoju HIV/AIDS przy zmieniających się współczynnikach, przy ustalonych liczbach $X(0)=100$, $Y(0)=1$, $A(0)=0$:

- a) Załóżmy, że naturalna śmiertelność jest bardzo niska $\mu = 0.00001$, nie ma dopływu podatnych ludzi $B=0$. Transmisja choroby przy kontakcie, konwersja z grupy $Y$ do $A$ i przenoszenie  HIV/AIDS wynoszą $100 \%$, oraz, że śmiertelność związana z HIV/AIDS wynosi $d=0.005$ i średnia liczba partnerów to $c=0.13$. Zauważmy, że nie dochodzi do wybuchu epidemii:

```python
print(HIV(100,1,0,0,0.00001,1,0.13,0.005,1,1))
```

- b) zmieniamy średnią liczbę partnerów $c=5$, aby spełnić warunek wybuchu epidemii i współczynnik konwersji ustawiamy na $50 \%$:

```python
print(HIV(99,1,0,0,0.00001,1,5,0.005,1,0.5))
```

- c) wiemy, że współczynik konwersji z grupy $Y$ do $A$ wynosi około $25 \%$:

```python
print(HIV(99,1,0,0,0.00001,1,5,0.005,1,0.25))
```

- d) aby bardziej urzeczywistnić nasz model zwiększmy grupę, w pierwszym przypadku nie zmieniamy współczyników (mamy do czynienia z wymarciem całej populacji), a w drugim dokładamy stały wspołczynnik $B=10$ (co skutkuje stałym wzrostem całej populacji): 

```python
print(HIV(999,1,0,0,0.001,1,5,0.015,1,0.25))
print(HIV(999,1,0,10,0.001,1,5,0.015,1,0.25))
```

- e) zmienijąc współcznynniki na bardziej rzeczywiste, zwiekszając współczynnik śmiertelności nie związanej z HIV/AIDS na $3\%$, zmniejszając procent osób zarażonych przenoszących HIV/AIDS na $65\%$ , zmieniająć procent prawdopodobieństwa zakażenia się po kontakcie z osobą zakażoną na $5%$ i dokładając stałą liczbę dochodzących ludzi $B=2$:  

```python
print(HIV(99,1,0,2,0.03,0.5,5,0.005,0.65,0.25))
```

Jak widzimy głównie, model epidemii HIV/AIDS zależy od charakterystyki populacji i jej specyfiki. Istnieje możliwość wyginięcia całej grupy na skutek, śmiertleności choroby lub umieralności z nią nie związanej, jak i mamy możliwość kompletnego braku wybuchu epidemii. Posiadając dokładniejsze dane, możemy przeprowadzić symulacje, dla danego społeczeństwa.
Dla przykładu, według artykułu z $4$ grudnia $2019$ roku z portalu "Lubuskie Warte zachodu" pod tytułem "Obchody Światowego Dnia Walki z AIDS", na terenie województwa lubuskiego od $1995$ roku doszło do zakażenia $787$ osób, $241$ osób zaraziło się AIDS, oraz śmierć w związku z chorobą poniosło $63$ ludzi. Ponieważ, nie mamy dokładniejszych danych o chorobie spróbujmy oszacować resztę współczuyników. Od $1995$ do $2019$ w Lubuskiem średni przyrost ludzi na rok wychodzi ujemny i wynosi on $B \approx  -141$. Przyjmijmy, że naturalna śmiertelność to około $3\%$, wyliczając przybliżoną śmiertelność związaną z chorobą otrzymujemy wynik w przybliżeniu $0.026$. Biorąc pod uwagę, że średnia ilość partnerów to około pięciu i przyjmując, że każdy zakażony przenosi chorbę i prawdopodobieństwo transmisji choroby przy kontakcie wynosi $100\%$ otrzymujemy:  

```python
HIV(1011600,787,241,-141,0.03,1,5,0.026,1,0.2)
```

Oczywiście, nasz model nie bierze pod uwagę dynamiki rozwoju populacji, dobrze wiemy, że przyrost $B$ powinien się zmieniać z roku, na rok. W taki sposob otrzymujemy symulacje, jak może na terenie wojęwództwa Lubuskiego, zachowywać się choroba HIV/AIDS.
